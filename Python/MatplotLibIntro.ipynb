```python
#plotting graphs using matplotlib

import matplotlib.pyplot as plt
import numpy as np
```

```python
X=[1,2,3]
Y=[2,4,6]
plt.plot(X,Y)
plt.scatter(X,Y)
plt.show()
```

```python
X2=[7,8,9]
Y2=[10,11,12]
plt.plot(X,Y)
plt.plot(X2,Y2)
plt.scatter(X,Y,color='b',linestyle='--')
plt.scatter(X2,Y2)
plt.show()
```

```python
np.array([1,2,3,4,5])
x=np.arange(0,5,0.01)
y=x**3
plt.plot(x,y)
plt.show()
```

```python
sizes=[3,4,6,2]
color=["blue","purple","pink","red"]
Label=["A","B","C","D"]
explode=[0.2,0,0,0]
plt.title("Split among the clusters")
plt.pie(sizes,colors=color,labels=Label)
plt.pie(sizes,colors=color,labels=Label,autopct="%.2f%%",explode=explode)
plt.show()
```
